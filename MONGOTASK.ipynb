```python
 #4.Find the total and average of the exam, quiz and homework and store them in a separate collection.
x = mycol.aggregate([{'$unwind':'$scores'},{'$group':{'_id':{'type':"$scores.type"},"total:":{"$sum":"$scores.score"},"average:":{"$avg":"$scores.score"}}}])
for i in x:
  print(i)
  a1.insert_one(i)
```

```python
#2. Find students who scored below average in the exam and pass mark is 40%?
B1 = mydb["bwavg"]
```

```python
n1 = mycol.aggregate([{'$unwind':'$scores'},{'$group':{'_id':'$_id','name':{'$first':'$name'},'maxisco':{'$sum':'$scores.score'}}},{'$addFields':{'prct':{'$divide':['$maxisco',3]}}}])
for i in n1:
  if i["prct"]>=40 and i["prct"]<=50:
     B1.insert_one(i)

```

```python
for i in B1.find({},{'_id':0,'name':1,'prct':1}):
  print(i)
```

```python
#1)      Find the student name who scored maximum scores in all (exam, quiz and homework)?

c1 = mycol.aggregate([{'$unwind':'$scores'},{'$group':{'_id':'$_id','name':{'$first':'$name'},'maxisco':{'$sum':'$scores.score'} }}])
max = 0
for i in c1:
  if i['maxisco']>max:
    max = i['maxisco']
    name = i['name']
print(name,max)
```

```python
#3.Find students who scored below pass mark and assigned them as fail, and above pass mark as pass in all the categories.

d1 = mycol.aggregate([{'$unwind':'$scores'},{'$group':{"_id":"$_id","name":{"$first":"$name"},"maxisco":{'$sum':"$scores.score"}}},
                      {'$addFields':{"percent":{'$divide':["$maxisco",3]}}},
                      {'$addFields':{"result":{"$cond":{'if' :{"$gte":["$prcnt",40]},'then':"Pass",'else':"Fail"}}}}])
for i in d1:
  print(i['name'],i['result'])
```

```python
#7)      Create a new collection which consists of students who scored above pass mark in all the categories.
e1 = mycol.aggregate([{'$unwind':'$scores'},
                              {'$group':{"_id":"$_id","name":{"$first":"$name"},"max_sco":{'$sum':"$scores.score"}}},
                              {'$addFields':{"prcnt":{'$divide':["$max_sco",3]}}},
                              {'$addFields':{"result":{"$cond":{'if' :{"$gte":["$prcnt",40]},'then':"Pass",'else':"Fail"}}}}])
        
for i in e1:
   if i['result']== 'Pass':
     w.insert_one(i)
    
```
